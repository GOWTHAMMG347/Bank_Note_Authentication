```python
import numpy as np
from numpy import genfromtxt
```

```python
data=genfromtxt('C:/Users/GOWTHAM/Downloads/bank_note_data.txt',delimiter=',')
```

```python
data
```

```python
labels=data[:,4]
```

```python
labels
```

```python
features=data[:,0:4]
```

```python
features
```

```python
x=features
y=labels
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_object=MinMaxScaler()
```

```python
scaler_object.fit(x_train)
```

```python
scaled_x_train=scaler_object.transform(x_train)
```

```python
scaled_x_test=scaler_object.transform(x_test)
```

```python
scaled_x_train.max()
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential()

model.add(Dense(4,input_dim=4,activation='relu'))

model.add(Dense(4,activation='relu'))

model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(scaled_x_train,y_train,epochs=50,verbose=2)
```

```python
predict=model.predict(scaled_x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
predict_labels=(predict>0.5).astype(int)

confusion_matrix(y_test,predict_labels)
```

```python
print(classification_report(y_test,predict_labels))
```

```python
model.save('bank_note_autenticate_model.keras')
```
